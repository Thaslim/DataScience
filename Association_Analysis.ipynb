```
import numpy as np 
import pandas as pd 
from mlxtend.frequent_patterns import apriori, association_rules
```

```
# load data into a pandas dataframe
data = pd.read_excel('Online Retail.xlsx') 
```

```
# show the top 10 rows
data.head(10)
```

```
# What are the columns of our data?
data.columns
```

```
# What are the different regions that the transactions are from? (Countries)
data.Country.unique()
```

```
data.isnull().sum()
```

```
# Stripping extra spaces in the description 
data['Description'].str.strip().str.rstrip()
  
# Dropping the rows without any invoice number 
data.InvoiceNo.dropna(inplace=True)
  
# Dropping all transactions which were done on credit 
indexNames = data[data['InvoiceNo'].str.contains('C')==True].index
```

```
indexNames
```

```
data.drop(indexNames,inplace=True)
```

```
data
```

```
# Splitting the data according to the region of transaction
# select data rows by country designated below
# use groupby('InvoiceNo', 'Description')['Quantity']
# sum data
# fillna(0)
# set_index('Invoice Number)

# Transactions done in France 
basket_France = (data[data['Country'] =="France"].groupby(['InvoiceNo', 'Description'])['Quantity']).sum().unstack().reset_index().fillna(0).set_index('InvoiceNo')
  
```

```
basket_France
```

```
# Transactions done in the United Kingdom 
basket_UK = (data[data['Country'] =="United Kingdom"].groupby(['InvoiceNo', 'Description'])['Quantity']).sum().unstack().reset_index().fillna(0).set_index('InvoiceNo')

# Transactions done in Portugal 
basket_Por = (data[data['Country'] =="Portugal"].groupby(['InvoiceNo', 'Description'])['Quantity']).sum().unstack().reset_index().fillna(0).set_index('InvoiceNo')

# Transactions done in Swedan  
basket_Sweden = (data[data['Country'] =="Sweden"].groupby(['InvoiceNo', 'Description'])['Quantity']).sum().unstack().reset_index().fillna(0).set_index('InvoiceNo')
```

```
# Define a hot encoding function to make the data suitable  
# for the concerned libraries 
# value >= 1 -> 1 and similar for less than 0
def hot_encode(x): 
    if x <= 0:
        return 0
    if x >= 1:
        return 1

# Encoding the datasets 
# apply hot_encode function to data from above
basket_encoded = basket_France.applymap(hot_encode) 
basket_France = basket_encoded 
  
basket_encoded = basket_UK.applymap(hot_encode) 
basket_UK = basket_encoded 
  
basket_encoded = basket_Por.applymap(hot_encode) 
basket_Por = basket_encoded 
  
basket_encoded = basket_Sweden.applymap(hot_encode) 
basket_Sweden = basket_encoded 
```

**Market basket Analysis - France**

```
basket_France
```

```
# for each of the countries build a model
# view apriori documation
# min_support = 0.05
frq_items_france = apriori(basket_France, min_support=0.05, use_colnames=True)
```

```
# Collecting the inferred rules in a dataframe 
# use association_rules with metric='lift' and mininum threshold = 1
rules1 = association_rules(frq_items_france, metric="lift", min_threshold=1)
# given the rules sort by confidence & lift, in descending order
#Association rules for freq Itemset in france after sorting with confidence &lift
rules1.sort_values(['confidence', 'lift'], ascending=[False, False],inplace=True)
rules1.head()
```

From above table it is clear that ,paper cups and paper and plates are bought together in France.

**Market basket Analysis - United Kingdom**

```
#Association rules for freq Itemset in UK 
frq_items_uk = apriori(basket_UK, min_support=0.05, use_colnames=True)
rules2 = association_rules(frq_items_uk, metric="lift", min_threshold=1)
rules2.head()
```

From above results we can understand that the frequent itemset in Uk doesn't even have 0.05 min support. 
So, try with lesser values for min support

```
#find frequent items with min support 0.03
frq_items_uk = apriori(basket_UK, min_support=0.03, use_colnames=True)
```

```
rules2 = association_rules(frq_items_uk, metric="lift", min_threshold=1)
#Association rules for freq Itemset in UK after sorting with confidence &lift
rules2.sort_values(['confidence', 'lift'], ascending=[False, False],inplace=True)
rules2.head()
```

The people of united kingdom like to buy different colors of teacup and saucer together

**Market basket Analysis - Portugal**

```
#Association rules for freq Itemset in Portugal after sorting with confidence &lift
frq_items_Por = apriori(basket_Por, min_support=0.05, use_colnames=True)
rules3 = association_rules(frq_items_Por, metric="lift", min_threshold=1)
rules3.sort_values(['confidence', 'lift'], ascending=[False, False],inplace=True)
rules3.head()
```
